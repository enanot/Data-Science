Importar los datos del fichero adjunto n un dataframe pandas y calcular el total de ventas en dolares ("revenue") de los primeros 6 pedidos



```python
data = pd.read_csv('./data/mexican.tsv',sep='\t')
```

Seleccionamos informacion relevante

```python
new_df=data[['order_id','quantity','item_price']]
new_df = new_df[new_df.order_id < 7]
new_df
```

Definimos la funcion para quitar el dolar

```python
def dolar_to_float(x):
    return float(x[1:])
```

Aplicamos la funcion para quitar el dolar a new_df

```python
#new_df.item_price =new_df.item_price.apply(dolar_to_float)
new_df.item_price =new_df.item_price.apply(lambda x : float(x[1:]))
new_df
```

multiplicamos y sumamos y creamos una columna nueva

```python
col_sum = (new_df.quantity*new_df.item_price) 
new_df['Revenue']=col_sum

new_df
```

```python
revenue = sum(col_sum)
revenue
```
